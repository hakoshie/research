```python
import pandas as pd
```

```python
f_columns=["year","id","efficacy","rx_agg","rx_pro_agg","rx_pro_dom","otc_agg","otc_pro_agg","otc_pro_dom","self_agg","self_pro_agg","self_pro_dom"]
final_df=pd.DataFrame(columns=f_columns)
final_df
```

```python
def data_extract(final_df,names,header,index,year,drop=0):
    for it,name in enumerate(names):
        if drop==0:
            data=pd.read_excel(name,header=None,names=header)
        else: 
            data=pd.read_excel(name,header=None)
            data.dropna(how="all",inplace=True)
            data.dropna(how="all",inplace=True,axis=1)
            data.columns=header
            data.reset_index(inplace=True)
        column_names=f_columns[1:]
        row_list=[data.loc[index,cname] if cname in header else 0 for cname in column_names]
        row_list.insert(0,year+it)
        final_df=pd.concat([final_df,pd.DataFrame(row_list,index=f_columns).T],ignore_index=False)
    return final_df
```

```python
# type 1
# header =[]
names=[
"2015-4.xls",
# "2015-5.xls",
"2016-4.xls",
# "2016-5.xls",
"2017-4.xls",
# "2017-5.xls",
"2018-4.xls",
# "2018-5.xls"
]
year=2015
header=["id","efficacy","num","agg_agg","agg_pro","agg_pro_dom","agg_pro_im","agg_im","rx_agg","rx_pro_agg","rx_pro_dom","rx_pro_im","rx_im","etc_agg","etc_pro_agg","etc_pro_dom","etc_pro_im","etc_im","id2","efficacy2","self_agg","self_pro_agg","self_pro_dom","self_pro_im","self_im","otc_agg","otc_pro_agg","otc_pro_dom","otc_pro_im","otc_im","hou_agg","hou_pro_agg","hou_pro_dom","hou_pro_im","hou_im"]
index=190
final_df=data_extract(final_df,names,header,index,year)
```
