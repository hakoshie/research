```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 画像ファイルが格納されているディレクトリのパス
image_dir = "./plots"

# 画像ファイルのリストを取得
image_paths = [os.path.join(image_dir, file) for file in os.listdir(image_dir) if file.endswith(".png")]
for i,x in enumerate(image_paths):
    print(i,x)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8), dpi=300)  # 画質を向上させるためにdpiを設定
plt.subplots_adjust(hspace=-0.7,wspace=0)  # hspaceを調整して真ん中の余白を変更
indices_to_display=[0,2,3,4,5,6]
# 指定したインデックスの画像を2x3のグリッドに配置して表示します。
for i, ax in enumerate(axes.flat):
    if i < len(indices_to_display):
        index = indices_to_display[i]
        if 0 <= index < len(image_paths):
            img = mpimg.imread(image_paths[index])
            ax.imshow(img)
            ax.axis('off')  # 軸を非表示にします。
        else:
            ax.axis('off')  # 空白のプロットを非表示にします。

# グラフ全体を表示します。
plt.tight_layout()
plt.show()
```
