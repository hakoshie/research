```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df_with_dummies=pd.read_csv("../../data/len3_drop_never.csv",encoding="shift-jis")
```

Estimation

```python
df_columns = df_with_dummies.columns # 列名をリストとして取得
import re

# 正規表現パターンを作成します
pattern = r'id|^elasped_(m[0-5]|[0-9]|1\d|20)$'

# 列名のリストを作成します
column_names = [col for col in df_columns if re.search(pattern, col)]

# 列名のリストを表示します
print(column_names)
rest_columns=["year","generic_per"]
column_names=rest_columns+column_names
# try:
#     column_names.remove("elasped_m1")
# except:
#     1

try:
    column_names.remove("elasped")
except:
    1
try:
    column_names.remove("id")
except:
    1
try:
    column_names.remove("id_114")
except:
    1
# column_names
```

```python
# 2008で正規化するかで結構estimateが変わります、1ぐらい
df_with_dummies["year"]=df_with_dummies["year"]-2008
```

```python
dep_name="rx_dom"
# 特定の列の値が0の行を削除します
# df_with_dummies = df_with_dummies[df_with_dummies[dep_name] > 0]

X = df_with_dummies[column_names]
y = df_with_dummies[dep_name]
# y_tr=np.log(y)
# model = sm.OLS(y, X)
model = sm.OLS(y, X)
# model= sm.OLS(np.log(y), sm.add_constant(X))
results = model.fit()
print(results.summary())
# 結果を文字列に変換します
results_string = results.summary2().as_text()

# 結果をテキストファイルに書き込みます
with open('result_rx.txt', 'w', encoding="utf-8", errors="ignore") as file:
    file.write(results_string)

```
