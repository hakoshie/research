```python
import pandas as pd
import os
os.chdir("../")
data_ke_3=pd.read_csv("./OTC/ingre_cate_len3.csv",index_col=0)
data_ke_2=pd.read_csv("./OTC/ingre_cate_len2.csv",index_col=0)
data_yaku=pd.read_csv("./yakuji/yearly/data_ship_whole.csv",encoding="shift-jis")
```

```python
data_yaku.columns
```

```python
# make lag columns only for shipment data
for col in ['stock_rx','stock_otc', 'stock_self']:
    print(col)
    data_yaku[f'{col}_lag']=data_yaku.groupby(["id"])[col].shift()
```

```python
data_yaku
```

```python
# len 3
df_3=pd.merge(data_yaku,data_ke_3,left_on="id",right_on="id",how="left")
# sort 
df_3=df_3[df_3["id"].astype(str).str.len()==3]
df_3.sort_values(by=["year","id","Approved"]).reset_index(drop=1).to_csv("./merged/len3_ship.csv",encoding="shift-jis")
df_3
```

```python
# len 2
df_2=pd.merge(data_yaku,data_ke_2,left_on="id",right_on="id",how="left")
df_2=df_2[df_2["id"].astype(str).str.len()==2]
df_2.sort_values(by=["year","id","Approved"]).reset_index(drop=1).to_csv("./merged/len2_ship.csv",encoding="shift-jis")
df_2
```
