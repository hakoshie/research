```python
final_columns=["year","month","id","efficacy","rx_agg","rx_pro_agg","rx_pro_dom","otc_agg","otc_pro_agg","otc_pro_dom","self_agg","self_pro_agg","self_pro_dom"]
```

```python
def data_extract(final_df,names,header,year,month,drop=0):
    for it,name in enumerate(names):
        if drop==0:
            data=pd.read_excel(name,header=None,names=header)
        else: 
            data=pd.read_excel(name,header=None)
            data.dropna(subset=3,inplace=True)
            data.dropna(thresh=200,inplace=True,axis=1)
            data.dropna(how="all",inplace=True,axis=1)
            try:
                data.columns=header
            except:
                print(data.head())
                print(name,len(data.columns))
                break
            data=data.replace(',','',regex=True)
            data.reset_index(inplace=True)
        column_names=final_columns[2:]
        nrow=data.shape[0]
        for i in range(nrow):
            row_list=[data.loc[i,cname] if cname in header else 0 for cname in column_names]
            row_list.insert(0,month+it-it//12*12)
            row_list.insert(0,year+it//12)
            
            # idで仕分けする場合
            # try:
            #     id=int(data.loc[i,"id"])
            # except:
            #     id=-1
            # # if isinstance(id, (int, float, complex)) and not isinstance(id, bool) and not math.isnan(id):
            # if type(id) is not str:
            #     if id>=0 and id<1000:
            #         final_df=pd.concat([final_df,pd.DataFrame(row_list,index=f_columns).T],ignore_index=False)
            # num でsubsetをとる
            try:
                num=int(data.loc[i,"num"])
                # print(i,num)
            except:
                # print(i,data.loc[i,"num"])
                num=-1
            if type(num) is not str:
                if num>=0:
                    final_df=pd.concat([final_df,pd.DataFrame(row_list,index=final_columns).T],ignore_index=False)
    return final_df
```

```python
# initialization
final_df=pd.DataFrame(columns=final_columns)
final_df
```
