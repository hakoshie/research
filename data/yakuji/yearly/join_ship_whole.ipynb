```python
import pandas as pd
import math
```

```python
f_columns=["year","id","efficacy","rx_dom","otc_dom","self_dom","stock_rx","stock_otc","stock_self"]
final_df=pd.DataFrame(columns=f_columns)
final_df
```

```python
def data_extract(final_df,names,header,index,year,drop=0):
    for it,name in enumerate(names):
        if drop==0:
            data=pd.read_excel(name,header=None,names=header)
        else: 
            data=pd.read_excel(name,header=None)
            data.dropna(how="all",inplace=True)
            data.dropna(how="all",inplace=True,axis=1)
            data.columns=header
            data.reset_index(inplace=True)
        column_names=f_columns[1:]
        nrow=data.shape[0]
        for i in range(nrow):
            row_list=[data.loc[i,cname] if cname in header else 0 for cname in column_names]
            row_list.insert(0,year+it)
            # idで仕分けする場合
            # try:
            #     id=int(data.loc[i,"id"])
            # except:
            #     id=-1
            # # if isinstance(id, (int, float, complex)) and not isinstance(id, bool) and not math.isnan(id):
            # if type(id) is not str:
            #     if id>=0 and id<1000:
            #         final_df=pd.concat([final_df,pd.DataFrame(row_list,index=f_columns).T],ignore_index=False)
            # num でsubsetをとる
            try:
                num=int(data.loc[i,"num"])
            except:
                num=-1
            if type(num) is not str:
                if num>=0:
                    final_df=pd.concat([final_df,pd.DataFrame(row_list,index=f_columns).T],ignore_index=False)
    return final_df
```
