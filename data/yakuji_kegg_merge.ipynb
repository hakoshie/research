```python
import pandas as pd
data_ke_3=pd.read_csv("./OTC/ingre_cate_len3.csv",index_col=0)
data_ke_2=pd.read_csv("./OTC/ingre_cate_len2.csv",index_col=0)
data_yaku=pd.read_csv("./yakuji/data_ship_whole.csv",encoding="shift-jis")
```

```python
# len 3
merged_df_3=pd.merge(data_yaku,data_ke_3,left_on="id",right_on="id",how="left")
merged_df_3.to_csv("merged_len3.csv",encoding="shift-jis")
merged_df_3
```

```python
# len 2
merged_df_2=pd.merge(data_yaku,data_ke_2,left_on="id",right_on="id",how="left")
merged_df_2.to_csv("merged_len2.csv",encoding="shift-jis")
merged_df_2
```
