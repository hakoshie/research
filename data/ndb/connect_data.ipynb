```python
def data_extract(final_df,names,year,in_hospital,oral,sheet=0):
    for it,name in enumerate(names):
        if sheet==0:
            df=pd.read_excel(name,header=2)
        else:
            df=pd.read_excel(name,header=2,sheet_name=sheet)
        df.dropna(axis=0,thresh=40,inplace=True)
        df.columns=df.columns.str.replace("\n","")
        df["薬効分類"].fillna(method="ffill",inplace=True)
        df["薬効分類名称"].fillna(method="ffill",inplace=True)
        df["year"]=year+it
        df["in_hospital"]=in_hospital
        df["oral"]=oral
        df=df.loc[:,final_columns]
        final_df=pd.concat([final_df,df],ignore_index=False)
    return final_df
```
