```python
import pandas as pd
import numpy as np
data=pd.read_csv("kegg_w_cate_split.csv",index_col=0)
data=data.drop(["url"],axis=1)
data=data.dropna(subset=["date"]).reset_index(drop=True)
data["release_year"]=pd.to_datetime(data["date"]).dt.year
data
```

```python
data_4=data.copy()
data_3=data.copy()
data_2=data.copy()
for i,ids in enumerate(data["id1"]):
    for j in range(4):
        column_name =f"id{j+1}"
        if not pd.isnull(data.loc[i,column_name]):
            data_4.loc[i,column_name]=str(data.loc[i,column_name])[:4]
            data_3.loc[i,column_name]=str(data.loc[i,column_name])[:3]
            data_2.loc[i,column_name]=str(data.loc[i,column_name])[:2]
            
```

```python
#melt 
data_4=data_4.melt(id_vars=["Approved","KEGG","Ingredient","date","release_year","成分"],value_name="id")
data_3=data_3.melt(id_vars=["Approved","KEGG","Ingredient","date","release_year","成分"],value_name="id")
data_2=data_2.melt(id_vars=["Approved","KEGG","Ingredient","date","release_year","成分"],value_name="id")
data_4=data_4[data_4["id"].notna()].reset_index(drop=1)
data_3=data_3[data_3["id"].notna()].reset_index(drop=1)
data_2=data_2[data_2["id"].notna()].reset_index(drop=1)
```

```python
data_4.to_csv("ingre_cate_len4.csv",encoding="shift-jis")
data_4
```
