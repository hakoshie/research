```python
import pandas as pd
data=pd.read_csv("approved_ingredients_kegg_with_category.csv",index_col=0)
data
```

```python
for i,ids in enumerate(data["id"]):
    if type(ids)==str:
        split_ids=ids.split()
        for j, id in enumerate(split_ids):
            column_name =f"id{j+1}"
            if column_name not in data.columns:
                data[column_name]=""
            data.at[i,column_name]=id
```

```python
data=data.drop("id",axis=1)
```
