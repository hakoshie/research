```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("../merged/len3_ndb_blp_DN_firm.csv",index_col=0)
data=data[data["メーカー名"]=="統一"]
```

```python
data.loc[:,["医薬品名","薬価基準収載医薬品コード"]].value_counts()
```

```python
code_set=set(data.loc[data["メーカー名"]=="統一","薬価基準収載医薬品コード"])
for elm in code_set:
    print(elm,set(data.loc[data["薬価基準収載医薬品コード"]==elm,"医薬品名"]))
```

```python
# 医薬品名に全角の「」を含むもの
data[data["医薬品名"].str.contains("「")].loc[:,["医薬品名","薬価基準収載医薬品コード"]].value_counts()
```

```python
# 医薬品名に全角の「」を含むもの
pd.set_option('display.max_rows', None)
data[~data["医薬品名"].str.contains("「")].loc[:,["医薬品名","薬価基準収載医薬品コード"]].value_counts()
```
