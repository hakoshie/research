```python
yakka=pd.DataFrame()
for i,name in enumerate(files):
    print(start_y+i//2,name)
    this_year = start_y+i//2
    last_year = (start_y+i//2)-1
    if this_year<=2015 and i%2==0:
        data_i=pd.read_excel(f"./datafiles/{name}",names=header0)
    elif this_year<=2015 and i%2==1:
        data_i=pd.read_excel(f"./datafiles/{name}",names=header1)
    else:
        data_i=pd.read_excel(f"./datafiles/{name}",names=header2)
    data_i['year']=start_y+i//2
    yakka = pd.concat([yakka,data_i],axis=0,ignore_index=True)
    # continue
    # 1年に1回のみ実行
    if i%2==1 and this_year>=2015:
        # 現在のDataFrameから年とIDの組み合わせを取得
        # yakka_2y=yakka[yakka['year']>=(last_year)]
        yakka_last=yakka[yakka['year']==(last_year)]
        yakka_this=yakka[yakka['year']==(this_year)]
        set_this=set(yakka_this["薬価基準収載医薬品コード"])
        set_last=set(yakka_last["薬価基準収載医薬品コード"])
        # existing_combinations = set(zip(yakka_2y['year'], yakka_2y["薬価基準収載医薬品コード"]))
        # this yearにだけある
        set_this_m_last=set_this-set_last
        # last year にだけある
        set_last_m_this=set_last-set_this

        # 去年に、今年のIDがない場合、今年のIDを持つ行をコピー、昨年に対応する行を作成
        new_rows = []
        for id in set_this_m_last:
            this_year_row = yakka[(yakka['year'] == this_year) & (yakka["薬価基準収載医薬品コード"] == id)]
            new_row = this_year_row.copy()
            new_row['year'] = last_year
            new_rows.append(new_row)
        if this_year in [2018,2019,2020]:
            # 今年に、last yearのIDがない場合、last yearのIDを持つ行をコピーしthis yearに対応する行を作成
            for id in set_last_m_this:
                    last_year_row = yakka[(yakka['year'] == last_year) & (yakka["薬価基準収載医薬品コード"] == id)]
                    new_row = last_year_row.copy()
                    new_row['year'] = this_year
                    new_rows.append(new_row)
        # 新しい行を現在のDataFrameに追加
        yakka = pd.concat([yakka] + new_rows,axis=0, ignore_index=True)
    
```

```python
import pandas as pd
import numpy as np
# yakka.to_csv("./datafiles/yakka_till2022.csv",index=False,encoding="utf-8-sig")
yakka=pd.read_csv("./datafiles/yakka_till2022.csv",encoding="utf-8-sig")
```

```python
yakka=yakka
yakka.sort_values(by=["薬価基準収載医薬品コード","year"],inplace=True)
yakka.reset_index(drop=True,inplace=True)
yakka.head(20)[["薬価","品名","year"]]
```

```python
yakka["next_price"]=yakka.groupby(by=["薬価基準収載医薬品コード"])["薬価"].shift(-1)
yakka["next_year"]=yakka.groupby(by=["薬価基準収載医薬品コード"])["year"].shift(-1)
R=0.02
yakka["wholesale_price"]=yakka["next_price"]-yakka["薬価"]*R
yakka["markup"]=yakka["薬価"]-yakka["wholesale_price"]
yakka.head(20)[["薬価","品名","year","next_price","wholesale_price","markup","next_year"]]
```

```python
# 2年に１度の薬価改定の時期を考慮する
for year in [2014,2016]:
    # 指定された年のデータが存在するIDを取得する
    df_t = yakka[(yakka["year"] == year) | (yakka["year"] == year+1)]
    df_t["cnt"]=df_t.groupby(by=["薬価基準収載医薬品コード"])["薬価"].transform("count")
    df_t=df_t[df_t["cnt"]==2]
    # print(df_t.head(5))
    df_t.loc[df_t["year"]==year,["wholesale_price","markup","next_price","next_year"]]=np.nan*4
    df_t=df_t.groupby(by=["薬価基準収載医薬品コード"]).bfill()
    yakka.loc[df_t.index,["wholesale_price","markup","next_price","next_year"]]=df_t[["wholesale_price","markup","next_price","next_year"]]
print(yakka.head(20)[["薬価","品名","year","next_price","wholesale_price","markup","next_year"]])
yakka=yakka[["year","wholesale_price","markup","薬価基準収載医薬品コード"]]
```

```python
yakka["wholesale_price"]=yakka.groupby(by=["薬価基準収載医薬品コード"])["wholesale_price"].bfill().ffill()
yakka["markup"]=yakka.groupby(by=["薬価基準収載医薬品コード"])["markup"].bfill().ffill()
```
